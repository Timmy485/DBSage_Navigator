```python
import pandas as pd
from io import StringIO

input_file_path = "../data/NETFLIX_REVIEWS.csv"
output_file_path = "../data/output.csv"

# Read the file in binary mode
with open(input_file_path, 'rb') as file:
    binary_content = file.read()

# Decode the content, replace errors
decoded_content = binary_content.decode('windows-1252', errors='replace')

# Convert the decoded content into a StringIO object and read it with Pandas
df = pd.read_csv(StringIO(decoded_content))

# Write CSV file with UTF-8 encoding
df.to_csv(output_file_path, index=False, encoding='utf-8')

print(f'Conversion completed. File saved to: {output_file_path}')
```

```python
input_file_path = "../data/output.csv"
output_file_path = "../data/output2.csv"

# Read the file with Windows-1252 encoding and replace errors
with open(input_file_path, 'r', encoding='windows-1252', errors='replace') as file:
    content = file.read()

# Write the content to a new file with UTF-8 encoding
with open(output_file_path, 'w', encoding='utf-8') as file:
    file.write(content)
```

```python
# Load a small sample of your dataset to infer column types
sample_data = pd.read_csv('../data/employee_records.csv', nrows=30)

# Define the table name
table_name = 'employee_data'

# Infer column types using pandas dtype mapping
dtype_mapping = {
    'int64': 'INTEGER',
    'float64': 'NUMERIC',
    'object': 'TEXT',
    'datetime64[ns]': 'TIMESTAMP',
}

# Generate the table definition SQL with double-quoted column names
table_definition = ',\n'.join([f'"{column}" {dtype_mapping[str(sample_data[column].dtype)]}' for column in sample_data.columns])

# Print the inferred data types for each column
# for column, data_type in zip(sample_data.columns, sample_data.dtypes):
#     print(f'{column}: {data_type}')

# Create the table creation SQL string
table_creation_sql = f"""
CREATE TABLE {table_name} (
    {table_definition}
);
"""

# Print the SQL string
print(table_creation_sql)
```
